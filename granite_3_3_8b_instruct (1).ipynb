```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="ibm-granite/granite-3.3-8b-instruct")
messages = [
    {"role": "user", "content": "Who are you?"},
]
pipe(messages)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("ibm-granite/granite-3.3-8b-instruct")
model = AutoModelForCausalLM.from_pretrained("ibm-granite/granite-3.3-8b-instruct")
messages = [
    {"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
import streamlit as st
import torch
from transformers import pipeline
import fitz  # PyMuPDF for PDF reading

# Load model
qa_model = pipeline("question-answering", model="deepset/roberta-base-squad2")

def extract_text_from_pdf(uploaded_file):
    doc = fitz.open(stream=uploaded_file.read(), filetype="pdf")
    return "".join([page.get_text() for page in doc])

def get_answer_from_text(context, question):
    result = qa_model(question=question, context=context)
    return result['answer']

def main():
    st.set_page_config(page_title="📘 Study Mate", page_icon="📖", layout="wide")

    # Sidebar
    with st.sidebar:
        st.image("https://cdn-icons-png.flaticon.com/512/2907/2907253.png", width=100)
        st.title("Study Mate")
        st.markdown("### Your AI-powered Study Buddy")
        st.markdown("Upload notes and get instant answers using AI. 🤖")

    st.markdown("""
        <style>
            .big-font {
                font-size:24px !important;
                font-weight: 600;
            }
            .stButton>button {
                background-color: #4CAF50;
                color: white;
                border-radius: 8px;
                font-weight: bold;
                height: 3em;
                width: 100%;
            }
            .stTextInput>div>div>input {
                border: 2px solid #4CAF50;
                border-radius: 10px;
            }
        </style>
    """, unsafe_allow_html=True)

    st.markdown("<p class='big-font'>📂 Upload a PDF and Ask Anything!</p>", unsafe_allow_html=True)

    uploaded_file = st.file_uploader("Upload your PDF notes", type=["pdf"])

    if uploaded_file is not None:
        with st.spinner("⏳ Extracting text from PDF..."):
            pdf_text = extract_text_from_pdf(uploaded_file)
        st.success("✅ Text extracted successfully!")

        question = st.text_input("🔍 Ask a question based on your notes:")

        if question:
            with st.spinner("🤔 Thinking..."):
                answer = get_answer_from_text(pdf_text, question)
            st.markdown("### 🧠 Answer:")
            st.success(answer)

if __name__ == '__main__':
    main()
```
